##Task 2 - Descriptive Statistics üìä

- **Calculate measures of central tendency (mean, median, mode) and measures of dispersion (variance, standard deviation) for key variables. Identify and handle outliers if necessary.**


**Deliverables**:

- **Descriptive statistics that reveal the central tendencies, variations, and potential outliers in the dataset.**:

  

```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# Calculate mean, median, and mode
mean_values = df[['Income', 'Age', 'MntWines']].mean()
median_values = df[['Income', 'Age', 'MntWines']].median()
mode_values = df[['Income', 'Age', 'MntWines']].mode().iloc[0]  # Use iloc[0] to get the first mode

# Display the calculated statistics
print("Mean Values:")
print(mean_values)
print("\nMedian Values:")
print(median_values)
print("\nMode Values:")
print(mode_values)


```

```python
# Calculate variance and standard deviation
variance_values = df[['Income', 'Age', 'MntWines']].var()
std_deviation_values = df[['Income', 'Age', 'MntWines']].std()

# Display variance and standard deviation
print("\nVariance Values:")
print(variance_values)
print("\nStandard Deviation Values:")
print(std_deviation_values)
```

```python
# Calculate z-scores for 'Income' to identify potential outliers
z_scores_income = stats.zscore(df['Income'])
outliers_income = df[abs(z_scores_income) > 3]  # Adjust the threshold as needed

# Visualize potential outliers for 'Income' using a box plot
plt.figure(figsize=(8, 6))
plt.boxplot(df['Income'])
plt.title("Box Plot for Income")
plt.ylabel("Income")
plt.show()

# Display potential outliers for 'Income'
print("\nPotential Outliers for 'Income':")
print(outliers_income)

```

## Task 3 - Probability Distributions üé≤

- **Identify variables that could follow specific probability distributions (e.g., Binomial, Normal). Calculate probabilities and expected values based on these distributions.**



**Deliverables**:

- **Determination of suitable probability distributions for relevant variables and corresponding calculated probabilities and expected values.**:

  

```python
from scipy.stats import binom

# Define parameters
n = 10  # Number of trials
p = 0.3  # Probability of success in each trial

# Calculate probabilities
prob_success_5 = binom.pmf(5, n, p)  # Probability of exactly 5 successes in 10 trials
expected_value = n * p  # Expected value

print(f"Probability of 5 successes: {prob_success_5}")
print(f"Expected value: {expected_value}")
```

```python
from scipy.stats import norm

# Define parameters
mu = 50  # Mean
sigma = 10  # Standard deviation

# Calculate probabilities
prob_x_less_than_45 = norm.cdf(45, mu, sigma)  # Probability that X is less than 45
expected_value = mu  # Expected value

print(f"Probability of X < 45: {prob_x_less_than_45}")
print(f"Expected value: {expected_value}")
```

```python
from sklearn.cluster import KMeans

# Select relevant features for segmentation (e.g., spending on different products)
X = df[['MntWines', 'MntMeatProducts', 'MntFishProducts']]

# Choose the number of clusters (K) based on your analysis or goals
k = 3

# Apply K-means clustering
kmeans = KMeans(n_clusters=k)
df['Cluster'] = kmeans.fit_predict(X)

# Visualize the clusters (e.g., scatter plot)
sns.scatterplot(data=df, x='MntWines', y='MntMeatProducts', hue='Cluster')
plt.xlabel('Spending on Wines')
plt.ylabel('Spending on Meat Products')
plt.title('Customer Segmentation')
plt.show()
```
